```python
import matplotlib.pyplot as plt
import nibabel as nib
import SimpleITK as sitk

from starter_code.utils import load_case

%matplotlib inline
```

```python
vol_path = '/data/kits19/data/case_00100/imaging.nii.gz'
```

## nibabel

```python
data_nib = nib.load(vol_path)
```

```python
vol_nib = data_nib.get_fdata()

print("volume shape:")
print(vol_nib.shape)
```

```python
spacing_nib = data_nib.affine

print("spacing (affine):")
print(spacing_nib)
```

```python
plt.imshow(vol_nib[0, :, :])
```

## SimpleITK

```python
data_itk = sitk.ReadImage(vol_path)
```

```python
vol_itk = sitk.GetArrayFromImage(data_itk).astype(float)

print("volume shape:")
print(vol_itk.shape)
```

```python
spacing_itk = data_itk.GetSpacing()

print("spacing:")
spacing_itk
```

```python
plt.imshow(vol_itk[:, :, 0])
```

```python
vol_itk_trans = vol_itk.transpose()

plt.imshow(vol_itk_trans[0, :, :])
```

```python
assert (vol_itk_trans == vol_nib).all()
```
